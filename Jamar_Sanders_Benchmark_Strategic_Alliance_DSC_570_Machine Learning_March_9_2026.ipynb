```python
import pandas as pd
import networkx as nx

# Load adjacency matrix dataset
file_path = "/workspaces/lineify.ai/SaveAsSpreadsheet.csv"
df = pd.read_csv(file_path, index_col=0)

# Convert adjacency matrix to graph
G = nx.from_pandas_adjacency(df)

print("Graph successfully created.\n")

# -----------------------------
# 1. Density of Ego Networks
# -----------------------------
ego_nvidia = nx.ego_graph(G, "Nvidia")
ego_honeywell = nx.ego_graph(G, "Honeywell")

density_nvidia = nx.density(ego_nvidia)
density_honeywell = nx.density(ego_honeywell)

print("Ego Network Density")
print("Nvidia Density:", density_nvidia)
print("Honeywell Density:", density_honeywell)


# -----------------------------
# 2. Degree Centrality
# -----------------------------
degree_centrality = nx.degree_centrality(G)

print("\nDegree Centrality")
for actor, value in degree_centrality.items():
    print(actor, ":", value)


# -----------------------------
# 3. Betweenness Centrality
# -----------------------------
betweenness = nx.betweenness_centrality(G)

print("\nBetweenness Centrality")
for actor, value in betweenness.items():
    print(actor, ":", value)


# -----------------------------
# 4. Closeness Centrality
# -----------------------------
closeness = nx.closeness_centrality(G)

print("\nCloseness Centrality")
for actor, value in closeness.items():
    print(actor, ":", value)


# -----------------------------
# 5. Prestige (In-Degree Centrality)
# -----------------------------
prestige = nx.in_degree_centrality(G) if G.is_directed() else nx.degree_centrality(G)

print("\nPrestige Scores")
for actor, value in prestige.items():
    print(actor, ":", value)


# -----------------------------
# 6. Compare Ego Density vs Centrality
# -----------------------------
print("\nComparison of Ego Actors")

print("Nvidia")
print("Density:", density_nvidia)
print("Centrality:", degree_centrality.get("Nvidia"))

print("\nHoneywell")
print("Density:", density_honeywell)
print("Centrality:", degree_centrality.get("Honeywell"))
```

# Impact of Informal Networks on the Power Structure

1. Informal Networks Concentrate Influence

Individuals or organizations with many connections in the network often gain more influence than their formal position might suggest. Actors with high degree centrality interact with many others and become key sources of information and coordination.

Impact:
Certain organizations become central hubs of influence.

Decision-making may shift toward those with the strongest informal relationships, not necessarily those with the highest formal authority.

For example, if Nvidia or Honeywell has numerous ties with other firms in the network, they may influence strategic decisions across multiple organizations.

2. Informal Brokers Gain Strategic Power

Actors positioned between different groups in the network can act as brokers, connecting otherwise unconnected clusters. These positions often correspond to high betweenness centrality.

Impact:
Brokers control the flow of information between groups.

They can influence negotiations, partnerships, and resource allocation.

In a technology–industrial network, companies connecting AI technology clusters with industrial automation clusters may gain significant strategic power.

3. Informal Networks Accelerate Information Flow

Information often spreads faster through informal relationships than through official communication channels.

Impact:
Organizations learn about opportunities or risks earlier.

Innovation spreads more quickly across connected entities.

In a corporate network, informal ties may allow companies to coordinate on new technologies or market strategies before formal partnerships are established.

4. Informal Networks Can Create Power Inequality

Because networks are rarely evenly distributed, some organizations occupy central positions, while others remain peripheral.

Impact:
Central organizations gain more influence over industry decisions.

Peripheral organizations may struggle to access resources or information.

This imbalance can reinforce existing power hierarchies within industries.

5. Informal Networks Strengthen Collaboration and Trust

Repeated interactions between organizations often lead to strong ties, which promote trust and cooperation.

Impact:
More efficient coordination on joint projects.

Reduced transaction and negotiation costs.

Strong informal relationships can make alliances more stable and productive.
